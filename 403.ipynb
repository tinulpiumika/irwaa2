```python
from urllib.request import Request, urlopen

req = Request(
    url = 'https://www.guessfactory.com/us/en/sale/women/view-all', 
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/192.168.56.1 Safari/537.36 QIHU 360SE'}
)
webpage = urlopen(req).read()
```

```python
from bs4 import BeautifulSoup 
#import re
import pandas as pd
```

```python
doc = BeautifulSoup(webpage,'html.parser')
```

```python
products = doc.find_all('div', {'class': 'product-grid__column'})
```

```python
products[0]
```

```python
#name
products[0].find('a', {'class': 'link product-tile__link'}).text.strip()
```

```python
#old price
products[0].find('span', {'class': 'price__strike-through'}).text.strip()
```

```python
#new price
newprice = products[0].find('span', {'class': 'price__red-color text-nowrap'}).text.strip().split('\n')
newprice[1]
```

```python
#discount
discount = products[0].find('span', {'class': 'price__red-color d-inline-block text-nowrap'}).text.strip().split(' ')
discount = discount[0].split('(')
discount[1]
```

```python
products[0].find('span', {'class': 'price__red-color d-inline-block text-nowrap'}).text.strip()
```

```python
#initialize the dataframe
df = pd.DataFrame(columns=['Product number', 'Product Name', 'New Price', 'Old Price', 'Discount'])
i=0
for prod in products:
    i=i+1
    productName = prod.find('a', {'class': 'link product-tile__link'}).text.strip()
    
    newprice = prod.find('span', {'class': 'price__red-color text-nowrap'}).text.strip().split('\n')
    #newPrice = newprice[1]
    
    oprice = prod.find('span', {'class': 'price__strike-through'}).text.strip()
    #oldPrice = f'${oprice[len(oprice)-1]}'

    discount = prod.find('span', {'class': 'price__red-color d-inline-block text-nowrap'}).text.strip().split(' ')
    discount = discount[0].split('(')
    
    #create row data 
    row = pd.DataFrame([[i,productName,newprice[1],oprice,discount[1]]],columns=['Product number', 'Product Name', 'New Price', 'Old Price', 'Discount'] )
    
    #concat dataframe
    df = pd.concat([df,row], ignore_index=True)
```
