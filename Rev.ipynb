```python
from bs4 import BeautifulSoup 
import pandas as pd
from urllib.request import Request, urlopen
```

```python
req = Request(
    url = 'https://www.guess.com/us/en/women/apparel/sale/denim', 
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/192.168.56.1 Safari/537.36 QIHU 360SE'}
)
webpage = urlopen(req).read()
```

```python
doc = BeautifulSoup(webpage, 'html.parser')
```

```python
products = products = doc.find_all('div', {'class': 'product-grid__column'})
#products[]
```

```python
df = pd.DataFrame(columns=['Product No', 'Product Name', 'Old Price', 'New Price', 'Discount'])
i = 0

#Create dataframe to scrape data
for prod in products:
  i=i+1
  #prodname
  prodname = prod.find('a', {'class': 'link product-tile__link'}).text.strip()

  #old price
  oldprice = prod.find('span', {'class': 'price__strike-through'}).text.strip()

  #new price
  newprice = prod.find('span', {'class': 'price__red-color text-nowrap'}).text.strip().split('\n')[1]

  #discount
  discount = prod.find('span', {'class': 'price__red-color d-inline-block text-nowrap'}).text.strip().split(' ')[0].split('(')[1]

  row = pd.DataFrame([[i,prodname,oldprice,newprice, discount]], columns=['Product No', 'Product Name', 'Old Price', 'New Price', 'Discount'])
  df = pd.concat([df,row], ignore_index=True)
```
